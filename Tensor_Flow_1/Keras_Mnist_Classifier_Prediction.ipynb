```
## Load the Model
model = tf.keras.models.load_model("mnist_classifier.h5")
```

```
## To predict we need some data so we again load the same test data from keras mnist dataset
_,(X_test,y_test) = tf.keras.datasets.mnist.load_data()
```

```
X_test.shape,y_test.shape
```

```
## Do Prediction
## Lets take 3rd picture
X_test[3:4].shape
```

```
## Actual target dataset for 1st picture
y_test[3]
```

```
## Prediction of 1st pictureset data
prediction = model.predict(X_test[3:4])
```

```
prediction
```

```
## As we see that it gave 10 output
## But we need one output. So we will do predicyion with highest Probability
import numpy as np
predict_num = np.argmax(prediction[0])
```

```
print(predict_num)
```
