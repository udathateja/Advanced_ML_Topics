## Boston house Price Predcition using TensorFlow 1.x and Keras

```
%tensorflow_version 1.x
```

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
from keras import datasets
```

```
tf.__version__
```

```
## This is only required for jupyter Notebook. Reset the Default Graph
tf.reset_default_graph()
```

```
(features,label),_ = datasets.boston_housing.load_data(test_split=0)
```

```
type(features),type(label)
```

```
features.shape,label.shape
```

### Build TensorFlow Graph

```
## Define Input Data Placeholder
X = tf.placeholder(shape = [None,13],dtype=tf.float32,name = "X-input")
### Lets Normalize the features
X_nor = tf.layers.batch_normalization(X,training=True)
y = tf.placeholder(shape= [None],dtype = tf.float32,name = "y-input") ## Actual Price
```

```
## Define weight and bias
W = tf.Variable(tf.zeros(shape=[13,1]),name = "Weights")
b = tf.Variable(tf.zeros(shape=[1]),name = "Bias")
```

```
## Define Prediction
#y_pred = tf.add(tf.matmul(X,W),b,name = "Output")
## Use Normalize Data
y_pred = tf.add(tf.matmul(X_nor,W),b,name = "Output")
```

```
## Define Loss Function
loss = tf.reduce_mean(tf.square(y_pred-y),name="Loss")
```

```
## Use Gradient Descent Optimizer to Minimize the loss
train_out = tf.train.GradientDescentOptimizer(0.02).minimize(loss)
```

#### Executing the Graph

```
## Lets Start the Graph Execution
sess = tf.Session()
## We need to first initialize the Variables so the we can use them
sess.run(tf.global_variables_initializer())
## Number of times of epochs to train the model
trainning_epochs = 500
```

```
## Claculate Trainning loss and trainning output
for epoch in range(trainning_epochs):
  _,train_loss = sess.run([train_out,loss],feed_dict={X:features,y:label})
  if epoch % 50 == 0:
    print("Trainning Loss at step :",epoch," is ",train_loss)
```

```
sess.close()
```

Use Keras to Build the Graph

```
features.shape,label.shape
```

```
## Initialize the Sequential Graph
model = keras.Sequential()
## Normalize the input data (features)
model.add(keras.layers.BatchNormalization(input_shape=(13,)))
## Add dense layer for prediction,weight and bias will automatically be declare by Keras
model.add(keras.layers.Dense(1))
## Lets compile the model which add optimizer and loss
model.compile(loss = "mse",optimizer="sgd",)
```

```
## Lets fit the model with input and label dataset
model.fit(features,label,epochs=500)
```

As we see that when we use the keras our loss is very less compare to tensorflow because we use different learning rate and initilization of weight and bias is randomply in keras.

```
## Lets save the model
model.save("Keras_boston_Prediction")
```
